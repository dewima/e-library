```python
# install library faker
!pip install faker
!pip install tabulate

#import library yang akan digunakan
from faker import Faker
from tabulate import tabulate
import csv

def user_table(n_user, is_print, output_csv=None):
    """
    Fungsi untuk membuat dummy data user table
    Header:
      - user_id
      - username
      - email
      - password
      - registration_date

    Arg:
      - n_user (int): jumlah user yang ingin dibuat
      - is_print (bool): jika True akan menampilkan hasil data
      - output_csv (str): nama file CSV untuk menyimpan data (opsional)

    Return:
      - table (dict): dictionary yang berisi data user
    """

    # Inisialisasi objek Faker
    FAKER = Faker('id_ID')

    # Buat table
    table = {}
    table["user_id"] = [i + 1 for i in range(n_user)]
    table["username"] = [FAKER.user_name() for _ in range(n_user)]
    table["email"] = [f"{name.lower()}@{FAKER.free_email_domain()}" for name in table["username"]]
    table["password"] = [FAKER.password() for _ in range(n_user)]  # Ganti dengan metode sesuai kebutuhan
    table["registration_date"] = [FAKER.date_this_decade() for _ in range(n_user)]  # Ganti dengan metode sesuai kebutuhan

    # Print table
    if is_print:
        show_data(table)

    # Save to CSV if output_csv is provided
    if output_csv:
        save_to_csv(table, output_csv)

    return table

def show_data(table):
    # Fungsi untuk menampilkan data menggunakan tabulate
    headers = list(table.keys())
    rows = zip(*table.values())
    table_str = tabulate(rows, headers=headers, tablefmt="pretty")
    print(table_str)

def save_to_csv(table, filename):
    # Fungsi untuk menyimpan data ke file CSV
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(table.keys())
        writer.writerows(zip(*table.values()))

# Membuat data table user dan menyimpannya ke file CSV
user_table_data = user_table(n_user=100, is_print=True, output_csv='user_data.csv')
```

```python
# Import library pandas
import pandas as pd

# Import file fk
library_data = pd.read_csv('library_data.csv')
category_data = pd.read_csv('category_data.csv')

from faker import Faker
import csv
import random

fake = Faker('id_ID')

# Fungsi untuk menghasilkan data dummy untuk tabel Book
def generate_book_data(num_records, libraries, categories):
    book_data = []
    for i in range(1, num_records + 1):
        library_id = random.choice(libraries)
        category_id = random.choice(categories)
        title = fake.catch_phrase()
        author = fake.name()
        available_qty = fake.random_int(min=1, max=50)
        book_data.append([i, library_id, category_id, title, author, available_qty])
    return book_data

# Fungsi untuk membaca data dari file CSV
def read_csv(file_name, column_index):
    with open(file_name, 'r') as file:
        reader = csv.reader(file, delimiter=';')  # Tentukan tanda pemisah yang sesuai
        next(reader)  # Skip header row
        data = [int(row[column_index]) for row in reader]
    return data


# Contoh data library dan category yang sudah ada dari file CSV
library_id = read_csv('library_data.csv', column_index=0)
category_id = read_csv('category_data.csv', column_index=0)

# Menghasilkan 250 data dummy untuk tabel Book
book_data = generate_book_data(250, library_id, category_id)

import csv

# Fungsi untuk menuliskan data ke dalam file CSV
def write_to_csv(data, file_name, header=None):
    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        if header:
            writer.writerow(header)
        writer.writerows(data)

# Menuliskan data dummy ke dalam file CSV
write_to_csv(book_data, 'book_data.csv',
             header=['book_id', 'library_id', 'category_id', 'title', 'author', 'available_qty'])

from tabulate import tabulate
import pandas as pd

# Fungsi untuk menampilkan data menggunakan tabulate
def show_data(table):
    # Konversi list ke DataFrame
    table_df = pd.DataFrame(table, columns=['book_id', 'library_id', 'category_id', 'title', 'author', 'available_qty'])

    # Hanya menampilkan kolom yang memiliki header
    headers = list(table_df.columns)
    non_empty_headers = [header for header in headers if header]
    table_str = tabulate(table_df[non_empty_headers], headers=non_empty_headers, tablefmt="pretty", showindex=False)  # Menambahkan showindex=False
    print(table_str)

# Fungsi untuk menyimpan data ke file CSV
def save_to_csv(table, filename):
    # Konversi list ke DataFrame
    table_df = pd.DataFrame(table, columns=['book_id', 'library_id', 'category_id', 'title', 'author', 'available_qty'])
    table_df.to_csv(filename, index=False)

# Menampilkan tabel book_data
show_data(book_data)

# Menyimpan tabel book_data ke file CSV
save_to_csv(book_data, 'book_data.csv')
```

```python
import csv
import random
from faker import Faker
from datetime import datetime, timedelta
from tabulate import tabulate

fake = Faker()

# Fungsi untuk menghasilkan data dummy untuk tabel Hold
def generate_hold_data(num_records, users, books):
    hold_data = []
    for i in range(1, num_records + 1):
        user_id = random.choice(users)
        book_id = random.choice(books)
        hold_date = fake.date_between(start_date='-30d', end_date='today')
        expiry_date = hold_date + timedelta(days=14)
        hold_data.append([i, user_id, book_id, hold_date, expiry_date])
    return hold_data

# Fungsi untuk menuliskan data ke dalam file CSV
def write_to_csv(data, file_name, header=None):
    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        if header:
            writer.writerow(header)
        writer.writerows(data)

# Fungsi untuk membaca data dari file CSV
def read_csv(file_name, column_index):
    with open(file_name, 'r') as file:
        reader = csv.reader(file, delimiter=';')  # Tentukan tanda pemisah yang sesuai
        next(reader)  # Skip header row
        data = [int(row[column_index]) for row in reader]
    return data

# Fungsi untuk menampilkan data dari file CSV dalam bentuk tabel
def show_data(file_name):
    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        data = [row for row in reader]
        headers = data[0]
        rows = data[1:]
        print(tabulate(rows, headers=headers, tablefmt='pretty'))

# Contoh data library dan category yang sudah ada dari file CSV
user_id = read_csv('library_data.csv', column_index=0)
book_id = read_csv('category_data.csv', column_index=0)

# Menghasilkan 245 data dummy untuk tabel Hold
hold_data = generate_hold_data(245, user_id, book_id)

# Menuliskan data dummy ke dalam file CSV
write_to_csv(hold_data, 'hold_data.csv',
             header=['hold_id', 'user_id', 'book_id', 'hold_date', 'expiry_date'])

# Menampilkan data dari file CSV dalam bentuk tabel
show_data('hold_data.csv')

```

```python

import csv
import random
from faker import Faker
from datetime import datetime, timedelta
from tabulate import tabulate

fake = Faker()

# Fungsi untuk menghasilkan data dummy untuk tabel Queue
def generate_queue_data(num_records, holds):
    queue_data = []
    book_hold_count = {}

    for i in range(1, num_records + 1):
        hold_id = random.choice(holds)

        # Menentukan nomor antrian
        if hold_id not in book_hold_count:
            book_hold_count[hold_id] = 1
        else:
            book_hold_count[hold_id] += 1

        queue_position = book_hold_count[hold_id]

        queue_data.append([i, hold_id, queue_position])  # Menghilangkan kolom user_id
    return queue_data

# Fungsi untuk menuliskan data ke dalam file CSV
def write_to_csv(data, file_name, header=None):
    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        if header:
            writer.writerow(header)
        writer.writerows(data)

def read_csv(file_name, column_index):
    with open(file_name, 'r') as file:
        reader = csv.reader(file, delimiter=',')  # Ganti delimiter menjadi koma
        next(reader)  # Skip header row
        data = [int(row[column_index]) for row in reader]
    return data

# Fungsi untuk menampilkan data dari file CSV dalam bentuk tabel
def show_data(file_name):
    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        data = [row for row in reader]
        headers = data[0]
        rows = data[1:]
        print(tabulate(rows, headers=headers, tablefmt='pretty'))

# Contoh data library dan category yang sudah ada dari file CSV
hold_id = read_csv('hold_data.csv', column_index=0)

# Menghasilkan 245 data dummy untuk tabel Queue
queue_data = generate_queue_data(245, hold_id)

# Menuliskan data dummy ke dalam file CSV
write_to_csv(queue_data, 'queue_data.csv',
             header=['queue_id', 'hold_id', 'queue_position'])  # Menghilangkan kolom user_id

# Menampilkan data dari file CSV dalam bentuk tabel
show_data('queue_data.csv')
```

```python
import csv
import random
from faker import Faker
from datetime import datetime, timedelta
from tabulate import tabulate

fake = Faker()

# Fungsi untuk menghasilkan data dummy untuk tabel Transaction
def generate_transaction_data(num_records, users):
    transaction_data = []
    for i in range(1, num_records + 1):
        user_id = random.choice(users)
        borrow_date = fake.date_between(start_date='-30d', end_date='today')
        due_date = borrow_date + timedelta(days=14)
        return_date = fake.date_between(start_date=borrow_date, end_date=due_date)  # Mencegah return_date melewati due_date

        # Menghilangkan kolom "status" dari data
        transaction_data.append([i, user_id, borrow_date, due_date, return_date])
    return transaction_data

# Fungsi untuk menuliskan data ke dalam file CSV
def write_to_csv(data, file_name, header=None):
    with open(file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        if header:
            writer.writerow(header)
        writer.writerows(data)

# Fungsi untuk menampilkan data dari file CSV dalam bentuk tabel
def show_data(file_name):
    with open(file_name, 'r') as file:
        reader = csv.reader(file)
        data = [row for row in reader]
        headers = data[0]
        rows = data[1:]
        print(tabulate(rows, headers=headers, tablefmt='pretty'))

# Contoh data yang sudah ada dari file CSV
user_id = read_csv('user_data.csv', column_index=0)

# Menghasilkan 300 data dummy untuk tabel Transaction
transaction_data = generate_transaction_data(300, user_id)

# Menuliskan data dummy ke dalam file CSV
write_to_csv(transaction_data, 'transaction_data.csv',
             header=['transaction_id', 'user_id', 'borrow_date', 'due_date', 'return_date'])

# Menampilkan data dari file CSV dalam bentuk tabel
show_data('transaction_data.csv')
```
